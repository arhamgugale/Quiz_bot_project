```python
from tkinter import *

root = Tk()
root.title("Quiz Bot")
root.geometry("600x600")

current_question_index = 0
score = 0
quiz_questions_data = []
selected_answer = StringVar()
quiz_results = [] 

gk_questions = [
    {"question": "What is the largest ocean in the world?",
     "options": ["Atlantic Ocean", "Indian Ocean", "Arctic Ocean", "Pacific Ocean"],
     "answer": "D"},
    {"question": "Who was the first person to walk on the Moon?",
     "options": ["Yuri Gagarin", "Neil Armstrong", "Buzz Aldrin", "Michael Collins"],
     "answer": "B"},
    {"question": 'Which country is known as the "Land of the Rising Sun"?',
     "options": ["China", "South Korea", "Japan", "Thailand"],
     "answer": "C"},
    {"question": "What is the capital city of Canada?",
     "options": ["Toronto", "Vancouver", "Ottawa", "Montreal"],
     "answer": "C"},
    {"question": 'Who wrote the play "Romeo and Juliet"?',
     "options": ["Charles Dickens", "William Shakespeare", "Jane Austen", "Mark Twain"],
     "answer": "B"}
]

gen_z_questions = [
    {"question": "What does 'rizz' mean?",
     "options": ["A type of drink", "Charisma/flirting skills", "Relaxation", "Running late"],
     "answer": "B"},
    {"question": "If someone says 'That’s cap', what do they mean?",
     "options": ["It’s the truth", "It’s a lie", "It’s cool", "It’s a hat"],
     "answer": "B"},
    {"question": "What does 'stan' mean?",
     "options": ["To strongly dislike", "To be a huge fan of", "To stand tall", "To ignore someone"],
     "answer": "B"},
    {"question": "If something is 'lit', how is it?",
     "options": ["Boring", "On fire literally", "Exciting/fun", "Dangerous"],
     "answer": "C"},
    {"question": "What does 'simp' mean?",
     "options": ["Someone who shows excessive affection/crushes hard", "A smart person", "A simple lifestyle", "Someone who lies a lot"],
     "answer": "A"}
]

fun_questions = [
    {"question": "What is the world’s longest river?",
     "options": ["Amazon", "Nile", "Yangtze", "Mississippi"],
     "answer": "A"},
    {"question": "Which country invented ice cream?",
     "options": ["Italy", "China", "USA", "France"],
     "answer": "B"},
    {"question": "Which is the only mammal that can fly?",
     "options": ["Squirrel", "Bat", "Owl", "Parrot"],
     "answer": "B"},
    {"question": "How many hearts does an octopus have?",
     "options": ["1", "2", "3", "4"],
     "answer": "C"},
    {"question": "Which fruit has its seeds on the outside?",
     "options": ["Banana", "Apple", "Strawberry", "Grape"],
     "answer": "C"}
]

science_questions = [
    {"question": "What gas do plants absorb from the atmosphere?",
     "options": ["Oxygen", "Carbon dioxide", "Hydrogen", "Nitrogen"],
     "answer": "B"},
    {"question": "What is the chemical symbol for gold?",
     "options": ["G", "Au", "Ag", "Go"],
     "answer": "B"},
    {"question": "Which organ in the human body produces insulin?",
     "options": ["Liver", "Pancreas", "Kidney", "Heart"],
     "answer": "B"},
    {"question": "Which part of the cell contains genetic material?",
     "options": ["Ribosome", "Cytoplasm", "Nucleus", "Mitochondria"],
     "answer": "C"},
    {"question": "What is the boiling point of water at sea level?",
     "options": ["90°C", "100°C", "110°C", "120°C"],
     "answer": "B"}
]
history_questions = [
    {
        "question": "Who was the first President of the United States?",
        "options": ["A. Abraham Lincoln", "B. Thomas Jefferson", "C. George Washington", "D. John Adams"],
        "answer": "C"
    },
    {
        "question": "In which year did World War II end?",
        "options": ["A. 1940", "B. 1942", "C. 1945", "D. 1950"],
        "answer": "C"
    },
    {
        "question": "Which ancient civilization built the pyramids?",
        "options": ["A. Romans", "B. Greeks", "C. Egyptians", "D. Chinese"],
        "answer": "C"
    },
    {
        "question": "Who was known as the “Iron Lady”?",
        "options": ["A. Angela Merkel", "B. Margaret Thatcher", "C. Indira Gandhi", "D. Queen Elizabeth II"],
        "answer": "B"
    },
    {
        "question": "Where did the Industrial Revolution begin?",
        "options": ["A. USA", "B. France", "C. Germany", "D. Great Britain"],
        "answer": "D"
    }
]
psychology_questions = [
    {
        "question": "Why do people make to-do lists?",
        "options": ["A. To decorate their desk", "B. To feel stressed", "C. To stay organized and reduce anxiety", "D. To impress others"],
        "answer": "C"
    },
    {
        "question": "What is it called when you feel better after talking to someone about your problems?",
        "options": ["A. Therapy", "B. Cooking", "C. Sleeping", "D. Arguing"],
        "answer": "A"
    },
    {
        "question": "Why do people form habits like brushing their teeth at the same time daily?",
        "options": ["A. Because it’s boring", "B. It's a healthy addiction", "C. Their brain likes routines", "D. To save time for school"],
        "answer": "C"
    },
    {
        "question": "What helps you stay motivated toward a goal?",
        "options": ["A. Getting no sleep", "B. Setting small rewards", "C. Complaining", "D. Ignoring it"],
        "answer": "B"
    },
    {
        "question": "Why do some people get nervous before public speaking?",
        "options": ["A. They hate talking", "B. It’s too easy", "C. Fear of being judged", "D. They forgot their name"],
        "answer": "C"
    }
]
business_questions = [
    {
        "question": "What does CEO stand for?",
        "options": ["A. Chief Economic Officer", "B. Central Executive Officer", "C. Chief Executive Officer", "D. Corporate Executive Organizer"],
        "answer": "C"
    },
    {
        "question": "Which document outlines a company’s goals and how it plans to achieve them?",
        "options": ["A. Resume", "B. Business Plan", "C. Invoice", "D. Memorandum"],
        "answer": "B"
    },
    {
        "question": "What is the term for the amount of money a business earns before expenses?",
        "options": ["A. Net income", "B. Gross profit", "C. Revenue", "D. Capital"],
        "answer": "B"
    },
    {
        "question": "In marketing, what does the '4Ps' stand for?",
        "options": ["A. Product, Price, Place, Promotion", "B. People, Process, Performance, Product", "C. Price, Planning, Promotion, Payment", "D. Product, Planning, Profit, Promotion"],
        "answer": "A"
    },
    {
        "question": "Which company is famous for the slogan “Just Do It”?",
        "options": ["A. Adidas", "B. Reebok", "C. Puma", "D. Nike"],
        "answer": "D"
    }
]
math_questions = [
    {
        "question": "What is the value of π (pi) to two decimal places?",
        "options": ["A. 3.14", "B. 3.13", "C. 3.15", "D. 3.12"],
        "answer": "A"
    },
    {
        "question": "What is the square root of 144?",
        "options": ["A. 12", "B. 14", "C. 16", "D. 10"],
        "answer": "A"
    },
    {
        "question": "What is 15% of 200?",
        "options": ["A. 25", "B. 30", "C. 35", "D. 40"],
        "answer": "B"
    },
    {
        "question": "Solve: 8 + (2 × 5)",
        "options": ["A. 18", "B. 20", "C. 28", "D. 30"],
        "answer": "A"
    },
    {
        "question": "What is the next prime number after 7?",
        "options": ["A. 9", "B. 10", "C. 11", "D. 13"],
        "answer": "C"
    }
]


quiz_map = {
    "G.K": gk_questions,
    "Gen Z Slangs": gen_z_questions,
    "Fun Quiz": fun_questions,
    "Science Quiz": science_questions,
    "History Quiz": history_questions,
    "Psychology Quiz": psychology_questions,
    "Business Quiz": business_questions,
    "Math Quiz": math_questions,
}

def clear_content():
    for widget in root.winfo_children():
        widget.destroy()

def show_result():
    clear_content()
    
    canvas = Canvas(root)
    canvas.pack(side="left", fill="both", expand=True)

    scrollbar = Scrollbar(root, orient="vertical", command=canvas.yview)
    scrollbar.pack(side="right", fill="y")
    
    canvas.configure(yscrollcommand=scrollbar.set)
    canvas.bind('<Configure>', lambda e: canvas.configure(scrollregion = canvas.bbox("all")))

    result_frame = Frame(canvas)
    canvas.create_window((0, 0), window=result_frame, anchor="nw")
    
    result_text = f"Quiz Complete! Your final score: {score} / {len(quiz_questions_data)}"
    Label(result_frame, text=result_text, font=("Arial", 16, "bold")).pack(pady=20)
    
    Label(result_frame, text="--- Detailed Results ---", font=("Arial", 12)).pack(pady=5)

    for i, result in enumerate(quiz_results):
        q_num = i + 1
        q_text = result["question"]
        user_ans = result["user_answer"]
        correct_ans = result["correct_answer"]
        is_correct = result["is_correct"]
        
        color = "green" if is_correct else "red"
        status = "CORRECT ✅" if is_correct else "WRONG ❌"

        Label(result_frame, 
              text=f"Q{q_num}: {q_text}", 
              font=("Arial", 11, "bold"), 
              wraplength=550, 
              justify="left").pack(pady=(10, 0), padx=10, anchor="w")
        
        Label(result_frame, 
              text=f"Status: {status}", 
              font=("Arial", 11, "italic"), 
              fg=color).pack(padx=10, anchor="w")

        if not is_correct:
            Label(result_frame, 
                  text=f"Your Answer: {user_ans}", 
                  font=("Arial", 10),
                  fg="red",
                  wraplength=550, 
                  justify="left").pack(padx=10, anchor="w")
        
        Label(result_frame, 
              text=f"Correct Answer: {correct_ans}", 
              font=("Arial", 10),
              fg="dark green",
              wraplength=550, 
              justify="left").pack(padx=10, anchor="w")
        
        Label(result_frame, text="").pack()

    Button(result_frame, text="Go Back to Menu", command=main_menu).pack(pady=20)


def next_question():
    global current_question_index, score

    current_q_data = quiz_questions_data[current_question_index]
    user_choice = selected_answer.get()
    
    correct_letter = current_q_data["answer"]
    
    answer_index = ord(correct_letter) - 65
    options = current_q_data["options"]
    correct_text = options[answer_index]
    
    is_correct = (user_choice == correct_text)
    
    if is_correct:
        score += 1

    quiz_results.append({
        "question": current_q_data["question"],
        "user_answer": user_choice if user_choice else "No Answer",
        "correct_answer": correct_text,
        "is_correct": is_correct
    })

    current_question_index += 1

    if current_question_index < len(quiz_questions_data):
        display_question()
    else:
        show_result()

def display_question():
    clear_content()
    q_data = quiz_questions_data[current_question_index]

    Label(root, text=f"Q{current_question_index + 1}. {q_data['question']}",
          font=("Arial", 12, "bold"), wraplength=550, justify="left").pack(pady=10)

    selected_answer.set(None)
    for opt in q_data["options"]:
        Radiobutton(root, text=opt, variable=selected_answer, value=opt, font=("Arial", 11)).pack(anchor="w", padx=20)

    Button(root, text="Next", command=next_question).pack(pady=20)


def run_quiz(questions):
    global current_question_index, score, quiz_questions_data, quiz_results
    current_question_index = 0
    score = 0
    quiz_questions_data = questions
    quiz_results = [] 
    display_question()


def main_menu():
    clear_content()
    Label(root, text="Select a Quiz Category", font=("Arial", 14, "bold")).pack(pady=20)
    
    quiz_frame = Frame(root)
    quiz_frame.pack(pady=10)

    quizzes = list(quiz_map.keys())
    num_cols = 2
    
    for i, quiz_name in enumerate(quizzes):
        Button(
            quiz_frame, 
            text=quiz_name, 
            command=lambda q=quiz_name: run_quiz(quiz_map[q]),
            width=15, 
            height=2,
            font=("Arial", 10, "bold"),
            bg="#e0e0e0"
        ).grid(row=i // num_cols, column=i % num_cols, padx=10, pady=5)

main_menu()
root.mainloop()
```
